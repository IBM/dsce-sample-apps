```python
import sqlite3, os, joblib
from datetime import datetime
from forecasting import get_forecast
from oos_prediction_new import (
    predict_oos_and_days_until_oos,
    check_oos_with_forecast,
    train_days_until_oos_regressor,
    train_oos_classifier
)
def get_forecast_values(start_date, sku_ids=None, store_ids=None, n=7):
    sku_ids = sku_ids or []
    store_ids = store_ids or []

    sales_df, campaigns_df, inventory_df = get_data()
    print("511",sales_df)
    propensity_df = pd.DataFrame()  # Placeholder for now
    propensity_df=pd.read_csv("data/propensity_output.csv") 
    sku_store_combos = inventory_df[['sku_id', 'store_id']].drop_duplicates()

    # Filtering based on sku/store
    if sku_ids and store_ids:
        sku_store_combos = sku_store_combos[
            sku_store_combos['sku_id'].isin(sku_ids) &
            sku_store_combos['store_id'].isin(store_ids)
        ]
    elif sku_ids:
        sku_store_combos = sku_store_combos[sku_store_combos['sku_id'].isin(sku_ids)]
    elif store_ids:
        sku_store_combos = sku_store_combos[sku_store_combos['store_id'].isin(store_ids)]

    all_preds, all_shap_inputs = [], []

    for row in sku_store_combos.itertuples(index=False):
        
        sku = row.sku_id
        store = row.store_id
        print(sku,store)
        preds, shap_input = get_forecast(
            inventory_df, sales_df, campaigns_df, propensity_df,
            sku_id=sku, start_date=start_date, store_id=store, n=n
        )
        all_preds.append(preds)
        all_shap_inputs.append(shap_input)

    preds_df = pd.concat(all_preds, ignore_index=True) if all_preds else pd.DataFrame()
    preds_df.rename(columns={"date": "sale_date"}, inplace=True)

    # OOS logic
    oof_df = check_oos_with_forecast(inventory_df, preds_df)
    if os.path.exists("models_forecast/oos_regressor.pkl"):
        reg = joblib.load("models_forecast/oos_regressor.pkl")
    else:
        reg, _, _, _ = train_days_until_oos_regressor(oof_df)

    if os.path.exists("models_forecast/oos_clf.pkl"):
        clf = joblib.load("models_forecast/oos_clf.pkl")
    else:
        clf, _, _, _ = train_oos_classifier(oof_df)
    df_predictions = predict_oos_and_days_until_oos(inventory_df, preds_df, clf, reg,n,sku_ids,store_ids)
    df_predictions['date_added']=datetime.now()
    df_predictions.rename(columns={"days_until_oos":"oos_days","will_go_oos":"oos"},inplace=True)
    df_predictions.to_sql("out_of_stock", get_connection("data/retail_analytics.db"), if_exists="append", index=False)
    preds_df['date_added']=datetime.now()
    preds_df.rename(columns={"forecasted_sales":"demand"},inplace=True)
    preds_df.to_sql("demand_forecast_new", get_connection("data/retail_analytics.db"), if_exists="append", index=False)
    return preds_df, df_predictions
```

```python
# out_of_stock_model.py
import pandas as pd
import numpy as np
from datetime import datetime
import joblib

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, f1_score, roc_auc_score,
    mean_squared_error, r2_score
)
from xgboost import XGBClassifier, XGBRegressor

# ========== LOAD DATA ==========
# inventory = pd.read_csv("inventory.csv", parse_dates=["date", "last_restock_date", "next_restock_date"])
# sales = pd.read_csv("sales.csv", parse_dates=["sale_date"])

# ========== FEATURE ENGINEERING ==========

# --- Aggregate sales per SKU/Store ---
sales_df["sale_date"] = pd.to_datetime(sales_df["sale_date"])
sales_agg = (
    sales_df.groupby(["sku_id", "store_id"])
    .agg(
        total_sales=("quantity", "sum"),
        first_sale_date=("sale_date", "min"),
        last_sale_date=("sale_date", "max"),
    )
    .reset_index()
)

sales_agg["avg_daily_sales"] = (
    sales_agg["total_sales"] /
    ((sales_agg["last_sale_date"] - sales_agg["first_sale_date"]).dt.days + 1)
)


# Merge with inventory
df = inventory_df.merge(sales_agg, on=["sku_id", "store_id"], how="left")
```

```python
df
```

```python

# --- Fill missing sales info ---
df["avg_daily_sales"].fillna(0, inplace=True)
df["total_sales"].fillna(0, inplace=True)

# --- Days since last restock ---
df["days_since_last_restock"] = (df["date"] - df["last_restock_date"]).dt.days

# --- Days until next restock ---
df["days_until_next_restock"] = (df["next_restock_date"] - df["date"]).dt.days

# --- Stock-to-threshold ratio ---
df["stock_ratio"] = df["current_stock"] / (df["reorder_threshold"] + 1)

# --- Cost/Price margin ---
df["margin"] = df["current_selling_price"] - df["cost_price"]

# --- Binary target: Will go OOS soon? ---
df["will_out_of_stock"] = (df["current_stock"] <= df["reorder_threshold"]).astype(int)

# --- Regression target: Days to OOS ---
df["days_to_oos"] = np.where(
    df["avg_daily_sales"] > 0,
    (df["current_stock"] / df["avg_daily_sales"]).round().astype(int),
    np.nan
)

# Drop rows without regression target
df_reg = df.dropna(subset=["days_to_oos"]).copy()

# ========== FEATURES ==========
feature_cols = [
    "current_stock", "reorder_threshold", "lead_time_days",
    "avg_daily_sales", "total_sales", "days_since_last_restock",
    "days_until_next_restock", "stock_ratio", "margin"
]

X_cls = df[feature_cols]
y_cls = df["will_out_of_stock"]

X_reg = df_reg[feature_cols]
y_reg = df_reg["days_to_oos"]

# ========== SPLIT DATA ==========
Xc_train, Xc_test, yc_train, yc_test = train_test_split(X_cls, y_cls, test_size=0.2, random_state=42, stratify=y_cls)
Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# ========== CLASSIFICATION MODEL ==========
clf = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    use_label_encoder=False,
    eval_metric="logloss"
)

clf.fit(Xc_train, yc_train)

yc_pred = clf.predict(Xc_test)
yc_prob = clf.predict_proba(Xc_test)[:, 1]

print("\n Classification Metrics:")
print("Accuracy:", accuracy_score(yc_test, yc_pred))
print("F1 Score:", f1_score(yc_test, yc_pred))
print("ROC AUC:", roc_auc_score(yc_test, yc_prob))

# ========== REGRESSION MODEL ==========
reg = XGBRegressor(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

reg.fit(Xr_train, yr_train)

yr_pred = reg.predict(Xr_test)

print("\n Regression Metrics:")
print("RMSE:", mean_squared_error(yr_test, yr_pred, squared=False))
print("R2 Score:", r2_score(yr_test, yr_pred))

# ========== SAVE MODELS ==========
joblib.dump(clf, "out_of_stock_classifier.pkl")
joblib.dump(reg, "days_to_oos_regressor.pkl")
print("\n Models saved: out_of_stock_classifier.pkl, days_to_oos_regressor.pkl")
```
